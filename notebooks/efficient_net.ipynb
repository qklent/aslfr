```python
from torchvision.models import efficientnet_v2_s as efficientnet
```

```python
cnn = efficientnet()
cnn
```

```python
def get_n_params(model):
    pp=0
    for p in list(model.parameters()):
        nn=1
        for s in list(p.size()):
            nn = nn*s
        pp += nn
    return pp
```

```python
get_n_params(cnn)
```
