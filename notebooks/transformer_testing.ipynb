```python
def reset_model_weights(layer):
    if hasattr(layer, 'reset_parameters'):
        print(f"weights were resetted for {layer}")
        layer.reset_parameters()
    else:
        if hasattr(layer, 'children'):
            for child in layer.children():
                reset_model_weights(child)
```

```python
def translate_text(source_text, max_length=50):
    # Encode the source text to tensor
    input_ids = tokenizer.encode(source_text, return_tensors="pt")
    
    model.eval()
    
    with torch.no_grad():
        # Generate text
        output = input_ids
        for _ in range(max_length):
            logits = model(input_ids, decoder_input_ids=output).logits[:, -1, :]
            next_token = torch.argmax(logits, dim=-1)
            print(logits.shape)
            print()
            print(next_token)
            output = torch.cat((output, next_token), dim=-1)
            print(output)
        
        # Decode the generated tensor to text
        generated_text = tokenizer.decode(output[0].tolist(), skip_special_tokens=True)
    
    return generated_text

# Translate text using the function
source_text = "Hello, how are you?"
translated_text = translate_text(source_text, max_length=30)
print(translated_text)
```
